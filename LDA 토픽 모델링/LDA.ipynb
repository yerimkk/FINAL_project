## 데이터 전처리

```python
import pandas as pd
import ast

# 불용어 파일 불러오기
stopwords_df = pd.read_csv('/content/drive/MyDrive/Final/stop_words_bareun.csv')


# 불용어 튜플 리스트 만들기
stopwords_tuples = [(row['word'], row['pos']) for _, row in stopwords_df.iterrows()]
```

```python
stopwords_df.info()
```

```python
# 원본 리뷰 파일 불러오기
reviews_df = pd.read_csv('/content/drive/MyDrive/Final/final_combined_labeled.csv')
```

```python
# 불용어 튜플 제거 함수
def remove_stopwords(review):
    # 리뷰를 문자열로 받았다고 가정하고, 이를 리스트로 변환
    review_tuples = ast.literal_eval(review)
    # 불용어에 해당하지 않는 튜플만 남기기
    cleaned_review = [tup for tup in review_tuples if tup not in stopwords_tuples]
    return str(cleaned_review)
```

```python
# 리뷰 컬럼에서 불용어 제거
reviews_df['stopwords_clean'] = reviews_df['bareun_pos'].apply(remove_stopwords)
```

### 품사 필터링

'NNG', 'NNP','MAG', 'VA', 'VV', 'XR'

```python
# 필요한 품사만 필터링
def filter_pos(tagged_text, pos_tags=['NNG', 'NNP','MAG', 'VA', 'VV', 'XR']):
    # tagged_text는 [(단어, 품사)] 리스트여야 함
    return [word for word, pos in eval(tagged_text) if pos in pos_tags]

reviews_df['all'] = reviews_df['stopwords_clean'].apply(filter_pos)
```

### +) 불용어 추가

```python
additional_stopwords = ['하','되','안','없','앱','다','좀','영어', '공부', 'ㅋ', '보','쓰','있','뜨', '학습','ㅎㅎ',
                        'ㅎ', 'ㅠ','ㅋㅋ','ㅜㅜ', '보', '쓰', '있', '뜨', '너무', '아주', '정말', '더', '많이', '진짜', '늘', '가', '받','나',
                        '어떻', '왜', '내', '많' , '정도' , '그냥', '깔',

                        ] # 추가적인 불용어 목록

# 추가 불용어 필터링 함수 (품사는 유지하고 불용어만 제거)
def remove_stopwords(column, stopwords=additional_stopwords):
    return [word for word in column if word not in stopwords]

# 추가 불용어 제거
reviews_df['filtered_all'] = reviews_df['all'].apply(lambda x: remove_stopwords(x))
```

## VOCA

```python
voca = reviews_df[reviews_df['app_name']=='voca']
speak = reviews_df[reviews_df['app_name']=='speak']
```

### 전체 리뷰

```python
import pandas as pd
from gensim import corpora
from gensim.models.ldamodel import LdaModel
from pprint import pprint
from gensim.models import CoherenceModel
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# 각 리뷰가 토큰화된 단어 리스트로 구성되어 있어야 함
processed_docs = voca['filtered_all']

# 단어 사전과 코퍼스 생성
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

# LDA 모델 학습
lda_model = LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15, random_state=42)

# 토픽 출력
pprint(lda_model.print_topics())

# 모델의 일관성 점수 계산
coherence_model_lda = CoherenceModel(model=lda_model, texts=processed_docs, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)

# 토픽 시각화
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```
