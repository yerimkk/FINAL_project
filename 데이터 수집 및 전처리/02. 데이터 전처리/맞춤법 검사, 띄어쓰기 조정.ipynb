## 1. 네이버 맞춤법 검사기 (hanspell)

```python
# 모듈 불러오기
import pandas as pd
from hanspell import spell_checker
```

```python
# 데이터 불러오기
df = pd.read_csv('final_data_all.csv')
df
```

```python
# 5000개씩 끊어서 새로운 데이터프레임 청크로 분할
chunk_size = 5000
chunks = [df[i:i + chunk_size] for i in range(0, df.shape[0], chunk_size)]

# 각 데이터프레임 확인
for idx, chunk in enumerate(chunks):
    print(f"DataFrame chunk {idx+1}:")
    print(chunk)
    print("\n")
```

```python
# 5000개씩 끊어서 새로운 데이터프레임 리스트로 분할
chunk_size = 5000
chunks = [df[i:i + chunk_size] for i in range(0, df.shape[0], chunk_size)]

# spell_checker를 사용하여 content_cleaned 컬럼을 처리하는 함수 만들기
def clean_text_with_spellchecker(text):
    try:
        result = spell_checker.check(text)
        corrected_text = result.checked
        return corrected_text, None
    except Exception as e:
        return text, str(e)

# 각 청크별로 spell check 수행 및 새로운 컬럼 추가
error_log = []  # 에러가 발생한 행의 인덱스를 저장할 리스트
processed_chunks = []

for chunk_idx, chunk in enumerate(chunks):
    print(f"Processing chunk {chunk_idx+1}/{len(chunks)}")
    new_content = []
    for idx, row in chunk.iterrows():
        corrected_text, error = clean_text_with_spellchecker(row['content_cleaned'])
        if error:
            error_log.append((idx, error))  # 에러 발생 시 인덱스와 에러 메시지 저장
        new_content.append(corrected_text)
    chunk['new_content_cleaned'] = new_content
    processed_chunks.append(chunk)
```

```python
# 결과 합치기
final_df = pd.concat(processed_chunks, ignore_index=True)
final_df.to_excel('final_df.xlsx')
```

## 2. 띄어쓰기 수정 (Pycospacing)

```python
# 모듈 불러오기
from pykospacing import Spacing

spacing = Spacing()
```

```python
# 비문자열 행 번호를 저장할 리스트를 초기화
non_string_indices = []

# 함수 적용 시 에러를 처리하기 위한 함수
def safe_spacing(x, index):
    try:
        return spacing(x)
    except Exception as e:
        non_string_indices.append(index)
        return x

# enumerate를 사용하여 각 요소와 인덱스를 함께 처리
df['new_content_cleaned'] = [safe_spacing(x, idx) for idx, x in enumerate(df['new_content_cleaned'])]

# 비문자열 값 또는 에러가 발생한 행 번호 출력
print('에러가 발생한 행 번호: ', non_string_indices)
```
