```python
import re
import emoji
from soynlp.normalizer import repeat_normalize

# 한글, 영어, 숫자 및 공백만 허용하는 패턴
pattern = re.compile(r'[^A-Za-z0-9ㄱ-ㅣ가-힣\s]+')

# URL 패턴
url_pattern = re.compile(
    r'https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)')

def clean(x):
    x = x.lower()  # 대문자를 소문자로 변환
    x = pattern.sub(' ', x)  # 한글, 영어, 숫자, 공백을 제외한 모든 문자 제거
    x = emoji.replace_emoji(x, replace='')  # 이모지 삭제
    x = url_pattern.sub('', x)  # URL 제거
    x = x.strip()  # 앞뒤 공백 제거
    x = repeat_normalize(x, num_repeats=2)  # 반복 문자 축소

    return x
```

```python
df_combined['content_cleaned'] = df_combined['content'].apply(lambda x: clean(str(x)))
```
