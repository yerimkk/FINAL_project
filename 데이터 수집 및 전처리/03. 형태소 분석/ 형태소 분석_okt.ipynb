```python
from konlpy.tag import Okt
import pandas as pd


# 형태소 분석기 초기화
okt = Okt()

# 형태소 분석 함수 정의
def morph_analysis(text):
    return okt.morphs(text)

# 데이터프레임의 'content' 열에 형태소 분석 적용
df['okt'] = df['content_cleaned'].apply(morph_analysis)
```

```python
# 결과 확인
df[['content', 'okt']].head()
```

```python
from konlpy.tag import Okt

# 형태소 분석기 초기화
okt = Okt()

# 품사 태깅 함수 정의
def pos_analysis(text):
    return okt.pos(text)

# 데이터프레임의 'content' 열에 품사 태깅 적용
df['okt_pos'] = df['content_cleaned'].apply(pos_analysis)

```

```python
# 결과 확인
df[['content', 'okt','okt_pos']].head()
```

```python
# CSV 파일로 저장
df.to_csv('/content/drive/MyDrive/review_data_okt.csv', index=False, encoding='utf-8-sig')
```
